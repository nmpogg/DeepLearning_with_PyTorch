Khi làm việc với mạng nơ-ron tích chập, kỹ thuật lazy init sẽ trở nên thuận tiện hơn nữa vì kích thước đầu vào (ví dụ: độ phân giải của hình ảnh) sẽ ảnh hưởng đến kích thước của mỗi lớp tiếp theo. Do đó, khả năng thiết lập các tham số mà không cần biết giá trị của kích thước tại thời điểm viết mã có thể đơn giản hóa đáng kể nhiệm vụ xác định và sau đó sửa đổi các mô hình của chúng ta

```python
import torch
from torch import nn
```

```python
net_init = nn.Sequential(nn.Linear(32, 256), nn.ReLU(), nn.Linear(256, 10))
```

```python
net_init[0].weight
```

```python
#lazy init
net = nn.Sequential(nn.LazyLinear(256), nn.ReLU(), nn.LazyLinear(10))
net[0].weight  # chưa khởi tạo nên không có trọng số
```

```python
X = torch.rand(2, 20)
net(X)

net[0].weight.shape
# nhận được số chiều input -> cắm vào và init lại trọng số
```
