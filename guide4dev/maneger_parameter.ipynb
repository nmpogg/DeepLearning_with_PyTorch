Khi chúng ta chuyển sang các kiến ​​trúc không chỉ dựa trên các lớp xếp chồng tiêu chuẩn, đôi khi chúng ta sẽ cần phải đi sâu vào việc khai báo và thao tác các tham số. Trong phần này, chúng ta sẽ đề cập đến các nội dung sau:

- Truy cập các tham số để gỡ lỗi, chẩn đoán và trực quan hóa.

- Chia sẻ các tham số giữa các thành phần khác nhau của mô hình.

```python
import torch
from torch import nn
```

```python
net = nn.Sequential(nn.LazyLinear(8),
                    nn.ReLU(),
                    nn.LazyLinear(1))

X = torch.rand(size=(2, 4))
print(X.shape)
print(X)
print(net(X).shape)
```

Khi một mô hình được định nghĩa thông qua lớp Sequential, trước tiên chúng ta có thể truy cập bất kỳ lớp nào bằng cách lập chỉ mục vào mô hình như thể đó là một danh sách. Các tham số của mỗi lớp được đặt thuận tiện trong thuộc tính của nó.

```python
net[2].state_dict()
# lưu ý: kích W có dạng (out_features, in_features) nên hơi ngược với ma trận input
```

```python
#truy cập bias
type(net[2].bias), net[2].bias.data
```

```python
# có thể truy cập vào gradient
net[2].weight.grad == None  # chưa tính đạo hàm vì chưa lan truyền ngược
```
