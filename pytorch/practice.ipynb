```python
import torch
torch.__version__
```

```python
print("CUDA available:", torch.cuda.is_available())
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
x_data
```

```python
import numpy as np
np_array = np.array(data)
x_np = torch.tensor(np_array)
x_np
```

```python
# Tensor mới giữ nguyên các thuộc tính (hình dạng, kiểu dữ liệu)
# của tensor đối số, trừ khi được ghi đè một cách rõ ràng
x_ones = torch.ones_like(x_data) # retains the properties of x_data
print(f"Ones Tensor: \n {x_ones} \n")

x_rand = torch.rand_like(x_data, dtype=torch.float) # overrides the datatype of x_data
print(f"Random Tensor: \n {x_rand} \n")
```

```python
#Khởi tạo tensor với shape định nghĩa trước

shape = (2, 3,)
rand_tensor = torch.rand(shape)
ones_tensor = torch.ones(shape)
zeros_tensor = torch.zeros(shape)

print(f"Random Tensor: \n {rand_tensor} \n")
print(f"Ones Tensor: \n {ones_tensor} \n")
print(f"Zeros Tensor: \n {zeros_tensor}")
```

```python
#các thuộc tính của tensor

tensor = torch.rand(3, 4)

print(f"Shape of tensor: {tensor.shape}")
print(f"Datatype of tensor: {tensor.dtype}")
print(f"Device tensor is stored on: {tensor.device}")
```

```python
# chuyển tensor vào gpu

if torch.cuda.is_available():
    tensor = tensor.to('cuda')
print(f"Device tensor is stored on: {tensor.device}")
```

```python
tensor = torch.ones(4, 4)
tensor[:,1] = 0 # duyệt tất cả các hàng, đặt giá trị các ptu cột thứ 2 bằng = 0
print(tensor)
```

```python
#ghép tensor

t1 = torch.cat([tensor, tensor, tensor], dim=1)
print(t1)

t2 = torch.stack([tensor, tensor, tensor], dim=1)
print(t2)
```

```python
#Nhân các tensor
# This computes the element-wise product
print(f"tensor.mul(tensor) \n {tensor.mul(tensor)} \n")
# Alternative syntax:
print(f"tensor * tensor \n {tensor * tensor}")
```

```python
# nhân ma trận

print(f"tensor.matmul(tensor.T) \n {tensor.matmul(tensor.T)} \n")
# Alternative syntax:
print(f"tensor @ tensor.T \n {tensor @ tensor.T}")
```

```python
# Các thao tác có hậu tố '_' được gọi là thao tác tại chỗ. Ví dụ: x.copy_(y), x.t_(), sẽ thay đổi x.

print(tensor, "\n")
tensor.add_(5)
print(tensor)

# Các thao tác tại chỗ giúp tiết kiệm bộ nhớ, nhưng có thể gây ra vấn đề khi tính toán
# đạo hàm do mất dữ liệu lịch sử ngay lập tức. Do đó, việc sử dụng chúng không được khuyến khích.
```

AutoGrad


```python
import torch
x = torch.arange(4.0, requires_grad=True)
x
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```
