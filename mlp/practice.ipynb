```python
import torch
from matplotlib import pyplot as plt
```

```python
# ReLU
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)
print(x)
print(y)
plt.plot(x.detach(), y.detach())
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.show()
```

đạo hàm của ReLU không khả vi tại x = 0 -> tại đó chỉ lấy đạo hàm bên trái

```python
y.backward(torch.ones_like(x))
plt.plot(x.detach(), x.grad)
plt.xlabel("x")
plt.ylabel("Gradient of ReLU")
plt.show()
```

```python
#sigmoid 
y = torch.sigmoid(x)
plt.plot(x.detach(), y.detach())
```

gọi h(x) là sigmoid function thì đạo hàm của sigmoid là h(x) * (1 - h(x))

```python
x.grad.data.zero_() # Clear out previous gradients
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(), x.grad)
```
